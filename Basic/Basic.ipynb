```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img=cv.imread("Ankush_image.jpeg")
```

```python
type(img)
```

## read image

```python
plt.imshow(img) ## RED green Blue
## open cv:  Blue green red
```

```python
fix=cv.cvtColor(img,cv.COLOR_BGR2RGB)
```

```python
plt.imshow(fix)
```

### change color in grayscale

```python
img_grey=cv.imread("Ankush_image.jpeg",cv.IMREAD_GRAYSCALE)
```

```python
img_grey.max()
```

```python
plt.imshow(img_grey,cmap="gray")
```

```python
plt.imshow(fix)
```

## resizing image

```python
fix.shape
```

```python
new=cv.resize(fix,(1000,400))
```

```python
plt.imshow(new)
```

```python
new_image.shape
```
